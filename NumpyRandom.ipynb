# Random Number

```python
# a random number is something that can not be predicted logically
# Pseudo random and True Random
# If there is a program to generate random number, it means the number can be predicted. Thus it is not truly a random number but a pseudo random
```

```python
from numpy import random
import numpy as np
```

```python
# GENERATE NUMBER NUMBER
x = random.randint(100)
print(x)

# GENERATE RANDOM FLOAT
x = random.rand()
print(x)

print("GENERATING RANDOM ARRAY")
x = random.randint(100, size=5)
print(x)

print("Generate a 2-D array with 3 rows, each row containing 5 random integers from 0 to 100:")
y = random.randint(100, size=(3,5))
print(y)


x = random.rand(5)
print(x)

# Generating Random Number from Array using choice() method
x = random.choice([3,5,7,9], size=(1,4,5))
print(x)

# The size returns the shape of the array
```

# RANDOM DATA DISTRIBUTION

```python
# data distribution is a list of all possible values and how often each value occurs. Such list are important when working with
# data science and statistics.
# A random distribution is a set of random numbers that follow a certain probability density function.
# Probability Density Function: A function that describes a continuous probability. i.e. probability of all values in an array.
```

```python
x = random.choice([3,5,7,9],p=[0.1,0.3,0.6,0.0],size=(3,5))
print(x)

x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))
print(x)
```

# Random Permutation

```python
# A permutationüòçrefers to an arrangement of elements. e.g. [3, 2, 1] is a permutation of [1, 2, 3] and vice-versa.

# The NumPy Random module provides two methods for this: shuffle() and permutation().
```

```python

# Shuffle makes changes to the original array
arr = np.array([1,2,3,4,5,6])
random.shuffle(arr)
print(arr)

# permutation returns the re-arranged array (and leaves the original array un-changed)  
arr1 = np.array([1,2,3,4,5])
print(random.permutation(arr1))
print(arr1)
```

# NORMAL (GAUSSIAN) DISTRIBUTION

```python
# The normal distibution fits the distribution of many events like IQ Scores, Heartbeat etc.
# it uses the random.normal() method 
# it has three parameters
# loc - (mean) - where the peak of the bell exist
# scale - (standard deviation) - how flat the graph distribution should be
# size - The sape of the returned array
```

```python
x = random.normal(size=(2,3))
print(x)
print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")

print("Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:")

y = random.normal(loc=1, scale=2, size=(2,3))
print(y)
sns.distplot(y, hist=False)
plt.show()

print("Note: The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve.")
```

# BINOMIAL DISTRIBUTION

```python
# Binomail distribution is a discrete distribution
# Discrete Distribution:The distribution is defined at separate set of events, e.g. a coin toss's result is discrete as it can 
# be only head or tails whereas height of people is continuous as it can be 170, 170.1, 170.11 and so on.
# it describes the outome of a binary scenario, e.g toss of a coin, it will either be head or tails
# it has 3 parameters
# n - number of trials
# p - probability of each occurrence of each trial
# size - the shape of the returned array
```

```python
x = random.binomial(n=5, p=0.5, size=11)
print(x)

print("Visualization of Binomial Distribution")
sns.distplot(x, hist=False)
plt.show()

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False, label="Binomial")
plt.show()
```

```python
print("The main difference is that normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.")
sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show()
```

# POISSON DISTRIBUTION

```python
# Poisson distibution is a discrete distribution. It estimates how many times an event can happen in a specific time
# Like if someone eats twices a day, what is the probability that he will eat thrice
# it has 2 parameters
# lam = rate or known number of occurence e.g 2 for the above
# size = the shape of the returned array
```

```python
x = random.poisson(lam=2, size=10)
print(x)
print("Visualization of poisson distribution")
sns.distplot(x, kde=False)
plt.show()
```

```python
print("The difference between Normal and poisson is that Normal distribution is continous whereas poisson is discrete.")
# But we can see that similar to binomial for a large enough poisson distribution it will become similar to normal distribution 
# with certain std dev and mean.")

sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')
plt.show()
```

```python
# Difference Between Poisson and Binomial Distribution
print("The difference is very subtle it is that, binomial distribution is for discrete trials, whereas poisson distribution is for continuous trials.")
# But for very large n and near-zero p binomial distribution is near identical to poisson distribution such that n * p is nearly equal to lam.")

sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show()
```

# UNIFORM DISTRIBUTION

```python
# It is used to describe probability where event has equal chances of occuring
# e.g generation of random numbers
# it has 3 parameters
# a - lower bound - default is 0.0
# b - upper bound - default is 1.0
# size - The shape of the returned array
```

```python
print("Create a 2x3 uniform distribution sample")
x = random.uniform(size=(2,3))
print(x)

print("")
print("Visualization of uniform distribution")
sns.distplot(x)
plt.show()
```

# LOGISTIC DISTRIBUTION

```python
# Logistic distribution is used to describe growth
# it is used extensively in machine learning in logistic regression, neural networks etc.
# it has 3 parameters:
# loc - (mean) - the peak is default 0
# scale - (standard deviation), the flatness of the disribution. default 1
# size - the shape of the returned array
```

```python

x = random.logistic(loc=1, scale=2, size=(2,3))
print(x)
print("")
sns.distplot(x, hist=False)
plt.show()
```

```python
print("Difference between Logistics and Normal Distribution")
# Both distributions are near identical, but logistic distribution has more area under the tails. 
# ie. It representage more possibility of occurence of an events further away from mean.
# For higher value of scale (standard deviation) the normal and logistic distributions are near identical apart from the peak.

sns.distplot(random.logistic(loc=1, scale=2, size=(2,3)), hist=False, label="Logistic")
sns.distplot(random.normal(loc=1, scale=2, size=(2,3)), hist=False, label="Normal")
plt.show()
```

# MULTINOMIAL DISTRIBUTION

```python
# Multinomial distribution is a generalization of the binomial distribution.
# It describes outcomes of multi-nomial scenarios unlike binomial where scenarios must be only one of two. e.g. Blood type of a population, dice roll outcome.
# it has 3 parameters:
# n - number of possible outcome (e.g 6 for dice roll).
# pvals - list of probability of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).
# size - the shape of the returned array
x = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(x)
```

# Exponential Distribution

```python
# Exponential distribution is used for describing time till next event e.g. failure/success etc.
# It has two parameters:
# scale - inverse of rate ( see lam in poisson distribution ) defaults to 1.0.
# size - The shape of the returned array.
x = random.exponential(scale=2, size=(2,3))
print(x)
print("")
print("Visualization of exponential Distribution")
sns.distplot(random.exponential(size=1000), hist=False)
plt.show()

print("Note:")
print("Poisson distribution deals with number of occurences of an event in a time period whereas exponential distribution deals with the time between these events.")
```

# Chi Square Distribution

```python
# Chi sqaure distribution is used as a basis to verify the hypothesis
# it has 2 parameters
# df - the degree of freedom
# size - the shape of the returned array
```

```python
print("Draw out a sample for chi squared distribution with degree of freedom 2 with size 2x3:")

x = random.chisquare(df=2, size=(2,3))
print(x)

sns.distplot(x)
plt.show()
```

# Rayleigh Distribution

```python
# Rayleigh distribution is used in signal processing.
# It has two parameters:
# scale - (standard deviation) decides how flat the distribution will be default 1.0).
# size - The shape of the returned array.
```

```python
x = random.rayleigh(scale=2, size=(2, 3))
print(x)

sns.distplot(random.rayleigh(size=1000), hist=False)
plt.show()
print("")

print("Similarity Between Rayleigh and Chi Square Distribution")
print("At unit stddev the and 2 degrees of freedom rayleigh and chi square represent the same distributions.")
```

# Pareto Distribution

```python
# A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome).

# it has 2 parameters
# a - shape parameter
# size - The hspae of the returned array

x = random.pareto(a=2, size=(2, 3))
print(x)

sns.distplot(random.pareto(a=2, size=1000), kde=False)
plt.show()
```

# Zipf Distribution

```python
# Zipf distritutions are used to sample data based on zipf's law.
# Zipf's Law: In a collection the nth common term is 1/n times of the most common term. E.g. 5th common word in english has occurs nearly 1/5th times as of the most used word.

# It has two parameters:
# a - distribution parameter.
# size - The shape of the returned array.

x = random.zipf(a=2, size=(2, 3))
print(x)

print("")
print("Sample 1000 points but plotting only ones with value < 10 for more meaningful chart.")
x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10], kde=False)
plt.show()
```
